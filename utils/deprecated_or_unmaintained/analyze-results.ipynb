```python
from ipylab import JupyterFrontEnd
from IPython.display import Markdown, display
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
from rich.console import Console
console = Console(force_jupyter=True)

# Path to all CSVs
folder_path = "strategy_comparision_summary/*.csv"
csv_files = sorted(glob.glob(folder_path))

console.print(f"Found {len(csv_files)} files.")

# %%
for file_path in csv_files:
    file_name = os.path.basename(file_path)
    console.print(f"Processing: {file_name}")
    df = pd.read_csv(file_path)
    profit_row = df.iloc[0, 1:].astype(float)
    top5_periods = profit_row.sort_values(ascending=False).head(5)    
    # print("\nTop 5 periods with highest profit_total_abs:")
    # display(top5_periods)
    selected_columns = ["Unnamed: 0"] + top5_periods.index.tolist()
    detailed_table = df[selected_columns]
    console.print("Detailed metrics for Top 5 periods:")
    display(detailed_table)
```
